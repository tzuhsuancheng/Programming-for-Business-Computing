```python
# Question 1

# given information
info = []
price = []
quantity = []

for i in range(3):
    _input = input()
    _input_list = _input.split(',')
    if i == 0:
        for j in _input_list:
            info.append(int(j))
    elif i == 1:
        for j in _input_list:
            price.append(int(j))
    elif i == 2:
        for j in _input_list:
            quantity.append(int(j))

n = int(info[0]) # get n
c = int(info[1]) # get cost


# initial value:
max_profit = -1
profit_list = []
optimal_q = -1

# create a dataframe for price and quantity
data = {'p':price, 'q':quantity}

# find the maximum profit
for i in range(len(data['p'])):
    profit = (data['p'][i] - c) * data['q'][i]
    profit_list.append(profit)
    
    if profit >= max_profit:
        max_profit = profit
        
# find the optimal q and p
for i in range(len(profit_list)):
    if profit_list[i] == max_profit:
        if data['q'][i] > optimal_q:
            optimal_q = data['q'][i]
            optimal_p = data['p'][i]
        
print(str(optimal_p) + ',' + str(max_profit))
```

```python
# question 2

# given information
info = [] # x, c1, c2, c3, c4, n
r = [] # weekly revenue

for i in range(2):
    _input = input()
    _input_list = _input.split(',')
    if i == 0:
        for j in _input_list:
            info.append(int(j))
    elif i == 1:
        for j in _input_list:
            r.append(int(j))


n = int(info[5]) # get n
x = int(info[0]) # get captial

# total cost of a day
tc1 = int(info[1]) + int(info[2]) #c1+c2 # mon, thu 
tc2 = int(info[1]) + int(info[2]) + int(info[3]) #c1+c2+c3 # tue, fri
tc3 = int(info[1]) + int(info[2]) + int(info[4]) #c1+c2+c4 # wed


bal_list = [tc1, tc2, tc3, tc1, tc2, r] # revenue or cost of a week/ day

j = 0
while x > 0 and j != n :
    for bal in bal_list:
        if bal == bal_list[5]: 
            x = x + bal_list[5][j]  # add the revenue only on Sun
            j = j + 1               # use j to count the weeks
            if x <=0 or j == n:
                break
        else:
            x = x - bal           # calculate the rest capital day by day
            if x <=0:
                break
                
print(str(j) + ',' + str(x))
```

```python
# Question 4 
# given information
info = [] # n: number of the customer, m: number of the masseur
time = [] # time of each person
revenue = [] # revenue of each person

# append the input data to list
for i in range(3):                    
    _input = input()
    _input_list = _input.split(',')
    if i == 0:
        for j in _input_list:
            info.append(int(j))
    elif i == 1:
        for j in _input_list:
            time.append(int(j))
    elif i == 2:
        for j in _input_list:
            revenue.append(int(j))
            
m = info[1] # number of the masseur
hr = [ 0 for i in range(m) ] # create a list to calculate the working time of each masseur


# create a nested list for revenue and time
tmp = []
time_rev = []
for i in range(len(time)):
    tmp = [time[i], revenue[i]]
    time_rev.append(tmp)

    
# sorted time and revenue
time_rev = sorted(time_rev, key = lambda x:x[1], reverse = 1)
time_rev = sorted(time_rev, key = lambda x:x[0])



# calculate the customers and revenue under the conditions
count_n = 0 # count the number of customer
total_r = 0 # count the total revenue

for t in time_rev:
    hr[0] = hr[0] + t[0]
    
    if hr[0] > 360:
        hr[0] = hr[0] - t[0] # if >360 minus the revenue back (skip that customer)
    else:
        hr.sort()  # sort the list of working time 
        total_r = total_r + t[1]
        count_n = count_n + 1

        

# output
print(str(count_n) + "," + str(total_r))
```

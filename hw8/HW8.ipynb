## 第一題

- pseudo code: <br/>
input x, n  <br/> 
起始值x  <br/> 
先對起始x平方後取第12~5位數，令他為x1，如果取n個 <br/>
第二個亂數則對x1平方後取12~5位數....重複n次 <br/>
最後輸出的亂數會有n個 <br/>

- debug: <br/>
lstrip("0") can remove only leading 0, but if using strip("0"), you would remove ALL zero. Be careful!

```python
# Setup
_input = [int(i) for i in input().split(",")]
x = _input[0]
n = _input[1]

random = []


# Main
for i in range(0, n):
    tmp = str(x ** 2)
    
    if len(tmp) < 5:
        x = "0"
        
    else:
        tmp = tmp[::-1] # reverse the string
        
        if len (tmp) < 12:
            x = tmp[4:len(tmp)]
        else:
            x = tmp[4:12]
            
        x = x[::-1] # need to reverse back 

        # get rid of leading 0    
        if x[0] == "0":
            x = x.lstrip("0")
            if x == "":
                x = "0"

    random.append(x)
    x = int(x)
    

# print    
for i, j in enumerate(random):
    if i != (len(random) - 1) :
        print(j, end = ",")
    else:
        print(j, end = "")
```

## 第二題: 簡單的行事曆

- pseudo code: <br/>
input: Event：名稱、開始時間、結束時間、價值 <br/>
if 時間沒有衝突，排入schedule=[b,c,d]<br/>
if A 行程跟 b,c,d時間衝突:<br/>
    if A 價值 > b or c or d<br/>
        A 取代 b or c or d 
        
- class Event：
    - init：宣告變數名稱、開始時間、結束時間、價值
    - 
- class Schedule
    - init
    - def add_event 

```python
class Event(): # store the attributes of each event
    
    def __init__(self, name, t1, t2, value): # declare the frequent varialbes       
        self.name = name
        self.start = t1
        self.end = t2
        self.value = int(value)
        
        
    def strToTimeValue(self): # convert string Start/End time to seconds and store name, value to dictionary
        self.time = {}
        
        hr, mint, sec = self.start.split(':')        
        self.time['start'] = 3600 * int(hr) + 60 * int(mint) + int(sec)
        
        hr, mint, sec = self.end.split(':')        
        self.time['end'] = 3600 * int(hr) + 60 * int(mint) + int(sec)
        
        self.time['name'] = self.name
        self.time['value'] = self.value
        return self.time
        
        
class Schedule(): # check the conditions of each event
    
    def __init__(self):        
        self.events = [] # our schedule: store the events
        
        
    def sum_value(self, time_fail): # calculate the sum of value
        sum_value = 0   
        
        for t in time_fail:
            sum_value = sum_value + t['value']
            
        return sum_value
               
        
    def compare1(self, enew, eold): # compare the time of new events and the events which are scheduled
        self.time_fail = [] # store the event which has a clash to the new event
        
        for i, e in enumerate(eold):
            if enew['start'] >= e['end']:
                pass
            elif enew['end'] <= e['start']:
                pass
            else:
                self.time_fail.append(e)
                    
        return self.time_fail
                
        
    def compare2(self, enew, eold): # comparet the value of new event and the sum of arranged event
        sum_value = self.sum_value(self.time_fail)
        
        if enew['value'] > sum_value:
            for i in self.time_fail:
                eold.remove(i)
        
            eold.append(enew)
            
        return eold

    
    def add_event(self, e): # add the new events to schedule        
        self.events.append(e)      
        return self.events

 
# Main
n = int(input())
i = 0
s = Schedule()

while i < n: 
    event = [i for i in input().split(",")]
    e = Event(event[0], event[1], event[2], event[3])
    t = e.strToTimeValue()
    
    if i == 0: # add the first event directly
        result = s.add_event(t)
        i = i + 1
        continue
        
    else:
        add = s.compare1(t, result)
        if add == 0:
            result = s.add_event(t)
        else:
            result = s.compare2(t, result)
            
    i = i + 1    
    
# print output
sum_value = 0   
for r in result:
    sum_value = sum_value + r['value']
    
print(str(len(result)) + ',' + str(sum_value))
```
